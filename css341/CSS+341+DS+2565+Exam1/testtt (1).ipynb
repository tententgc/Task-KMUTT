```python
import pandas as pd
import numpy as np
tx1 = pd.read_csv('./data2/tx2010.csv')
tx2 = pd.read_csv('./data2/tx2011.csv')
tx3 = pd.read_csv('./data2/tx2012.csv')
tx4 = pd.read_csv('./data2/tx2013.csv')
tx5 = pd.read_csv('./data2/tx2014.csv')
tx6 = pd.read_csv('./data2/tx2015.csv')
```

### problem 9

```python
all_tx = [tx1, tx2, tx3, tx4, tx5, tx6]
df_merge = pd.concat(all_tx)
dl = df_merge
numberOfCus = len(df_merge['customer_id'].unique())
dl['lastPrice'] = np.where(dl['discount'].isnull(), dl['ticket_price'], dl['ticket_price']*(1 - dl['discount'].str.rstrip("%").astype(float)/100) )
dl
```

### problem 10

```python
allCus = pd.read_csv('./data2/customers.csv')
allCus
```

```python
sumCus = dl.groupby(['customer_id'])[['lastPrice']].sum().sort_values(by=['lastPrice'], ascending = False)
top = sumCus.head(20)
topAndbot = top.append(sumCus.tail(20))
topAndbot
```

```python
show = pd.merge(topAndbot,allCus,on='customer_id')
show[['customer_id','lastPrice','address_zip', 'address_state']]
```

### problem 11

```python
mergeState = pd.merge(dl, allCus)
mergeState['year'] = mergeState['transaction_date'].apply(lambda x: pd.to_datetime(x).strftime('%Y/%m/%d')[0:4])
mergeState 
```

```python
state = mergeState.groupby(['address_state'])[['lastPrice']].sum()
state
```

```python
sumbystate = mergeState.groupby(['address_state', 'year'])[['lastPrice']].sum()
sumbystate
```
