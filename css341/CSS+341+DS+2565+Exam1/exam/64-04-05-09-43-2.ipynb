# Part 1


```python
import pandas as pd
import numpy as np

item_categories = pd.read_csv('../data1/item_categories.csv')
items = pd.read_csv('../data1/items.csv')
sales = pd.read_csv('../data1/sales.csv')
shops = pd.read_csv('../data1/shops.csv')
```

### Problem 1: ตัดรายการที่มีค่า item_cnt_day หรือ item_price น้อยกว่า 0


```python
print(sales.shape)
sales.drop(sales[sales['item_cnt_day'] < 0].index, inplace = True)
print("Drop item_cnt_day < 0: ",sales.shape)
sales.drop(sales[sales['item_price'] < 0].index, inplace = True)
print("Drop item_price < 0: ",sales.shape)
```

### Problem 2: แสดงค่าเฉลี่ย, Median และพิสัยระหว่างควอร์ไทล์ (Interquartile Range) ของข้อมูล item_price


```python
print("Avergare of item_price: ",sales['item_price'].mean())
print("Median of item_price: ",sales['item_price'].median())
print("IQR of item_price: ",(sales['item_price']).quantile(0.75) - (sales['item_price']).quantile(0.25))
```

### Problem 3: รวมข้อมูลในไฟล์items dataset เข้าไปใน sales dataset


```python
item_sales = items.merge(sales, on='item_id')
```

### Problem 4: เปลี่ยนชนิดข้อมูล date จาก string เป็นวันที่ date


```python
item_sales['date'] = pd.to_datetime(item_sales['date'], format='%d.%m.%Y')
```

### Problem 5: มีทั้งหมดกี่ items จากทุก shop รวมกัน


```python
len((item_sales['item_id']).unique())
```

### Problem 6: item ใดที่ขายได้มากที่สุดในแต่ละ shop โดยแสดง shop_id, item_id และจำนวนที่ขายได้สุงสุด 10 shops


```python
item_sales.groupby(['shop_id', 'item_id']).count().sort_values(by='item_cnt_day', ascending=False).head(10).apply('item_cnt_day').to_frame()
```

### Problem 7: ให้แสดงค่าเฉลี่ย (รวมทั้ง s.d., max, min) ของราคาขายสินค้าของแต่ละร้านค้า โดยแสดงเพียงร้านที่มีค่าเฉลี่ยสูงสุด 5 อันดับและต่ำสุด 5 อันดับ (เรียงลำดับจากมากไปน้อย)


```python
item_sales.groupby('shop_id').agg({'item_price': ['mean', 'std', 'max', 'min']}).sort_values(by=('item_price', 'mean'), ascending=False).head(5)
```

### Problem 8: ให้แสดงจำนวนสินค้าที่ขายได้รวมในแต่ละวันของสัปดาห์และ ยอดขายรวมในแต่ละวันของสัปดาห์โดยเอาเฉพาะปี 2013 และ 2014 และเทียบผลของสองปีนี้เคียงข้างกัน (ซ้ายเป็นของ 2013, ขวาเป็น 2014)


```python
item_sales['year'] = pd.DatetimeIndex(item_sales['date']).year
item_sales_2013 = item_sales[item_sales['year']==2013]
left = item_sales_2013.groupby(item_sales['date'].dt.day_name()).agg({'item_cnt_day': 'sum', 'item_price': 'sum'})
left
```

```python
items_sales_2014 = item_sales[item_sales['year']==2014]
right = items_sales_2014.groupby(item_sales['date'].dt.day_name()).agg({'item_cnt_day': 'sum', 'item_price': 'sum'})
right
```

### Problem 9: นำเข้า (load) ไฟล์ tx20xx ทั้ง 6 ไฟล์ จากนั้นรวมต่อเข้าด้วยกันเป็นข้อมูลเดียว และเตรียมข้อมูลให้เหมาะสม เพื่อใช้วิเเคราะห์และนำเสนอข้อมูลรวมทั้ง 6 ปีนี้ ในแงม่ ุมต่างๆ ในข้อที่Error! Reference source not found.


คิด lastPrice คือข้อมูล ticket_price ที่คิดหักส่วนลดแล้ว


```python
all_tx = [tx1, tx2, tx3, tx4, tx5, tx6]
df_merge = pd.concat(all_tx)
dl = df_merge
numberOfCus = len(df_merge['customer_id'].unique())
dl['lastPrice'] = np.where(dl['discount'].isnull(), dl['ticket_price'], dl['ticket_price']*(1 - dl['discount'].str.rstrip("%").astype(float)/100) )
dl
```

### Problem 10: ยอดขายรวม (หลังหักส่วนลด) ของลูกค้าแต่ละคน


#### โดยเรียงลำดับจากมากไปน้อย พร้อมระบุ zipcode และรัฐ (ตามหลัง) และแสดงเฉพาะลูกค้ายอดขาย top 20 และ bottom 20


```python
allCus = pd.read_csv('../data2/customers.csv')
allCus
```

```python
sumCus = dl.groupby(['customer_id'])[['lastPrice']].sum().sort_values(by=['lastPrice'], ascending = False)
top = sumCus.head(20)
topAndbot = top.append(sumCus.tail(20))
topAndbot
```

```python
show = pd.merge(topAndbot,allCus,on='customer_id')
show[['customer_id','lastPrice','address_zip', 'address_state']]
```

### Problem 11: ยอดขายรวม (หลังหักส่วนลด) ของแต่ละรัฐ


#### โดยที่สำหรับแต่ละรัฐให้แยกรายปี (เรียงปีจากน้อยไปมาก) และเรียงรฐัจากยอดขายมากไปน้อย (ให้แสดงยอดรวมของแต่ละรัฐด้วย)


```python
mergeState = pd.merge(dl, allCus)
mergeState['year'] = mergeState['transaction_date'].apply(lambda x: pd.to_datetime(x).strftime('%Y/%m/%d')[0:4])
mergeState
```

```python
df_agg = mergeState.groupby(['address_state', 'year'] ).agg({'lastPrice':sum})
g = df_agg['lastPrice'].groupby('address_state', group_keys=False)
res = g.apply(lambda x: x.sort_values(ascending=True))
c = res.to_frame()
c
```

```python
byState = c.pivot_table('lastPrice',['address_state'], 'year')
byState['sum'] = byState.sum(axis=1)
sortbyStateSum = byState.sort_values(by=['sum'], ascending = False)
sortbyStateSum
```

### Problem 12: จากข้อที่แล้ว ให้แสดงเฉพาะรัฐที่เป็น top 15 และ bottom 15


```python
sortbyStateSum
topState = sortbyStateSum.head(15)
topAndbotState = topState.append(sortbyStateSum.tail(15))
topAndbotState
```
