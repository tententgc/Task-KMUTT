```python
def BisectionInterest(M, L, m):
    def original_eq(x): return 12*M/x*(1-(1+x/12)**(-12*m))-L

    def is_same_sign(a, b): return np.sign(
        original_eq(a)) == np.sign(original_eq(b))
    xa = 0.0000001
    xb = 100.0
    xc = 0  # Ans of the interest rate
    epsilon = 1e-6
    if xa >= xb:
        print('a must < b')
    elif is_same_sign(xa, xb):
        print("Please choose new xa,xb")
    else:
        err = xb - xa
        for i in range(100):
            err /= 2
            xc = xa + err
            if err < epsilon:
                print('converged')
                break
            if is_same_sign(xa, xc):
                xa = xc
            else:
                xb = xc
            print(i, xc, err, original_eq(xc))
    print('{:.10f}%'.format(xc*100))
    return xc
```

```python
M = 600
L = 150000
m = 30
interestRate = BisectionInterest(M, L, m)
print(interestRate)
```
